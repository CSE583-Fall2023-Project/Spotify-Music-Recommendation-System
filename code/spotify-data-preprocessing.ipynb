```python
import pandas as pd
import hashlib
import ast
import random
```

```python
spotify_data = pd.read_csv("./data/data.csv")
spotify_data.head()
```

### Rename columns

```python
spotify_data = spotify_data.rename(columns={'id': 'song_id', 'artists': 'artist_name', 'name': 'song_name'})
spotify_data.head()
```

### Create new column artist_id

```python
def generate_artist_id(artist_name):
    """
    Generate a unique artist_id for each artist_name using a hash function.
    The artist_id will be an alphanumeric string.
    """
    # Using SHA256 hash function to generate a unique hash for each artist_name
    hash_object = hashlib.sha256(artist_name.encode())
    # Taking the first 10 characters to keep the ID reasonably short
    artist_id = hash_object.hexdigest()[:10]
    return artist_id

spotify_data['artist_id'] = spotify_data['artist_name'].apply(generate_artist_id)

print(spotify_data['artist_id'].nunique() == spotify_data['artist_name'].nunique())
```

### Create new column genre

```python
data_w_genres = pd.read_csv('./data/data_w_genres.csv')
data_w_genres.head()
```

```python
aggregated_genres = data_w_genres.groupby('artists')['genres'].apply(list).reset_index()
aggregated_genres
```

```python
# First, convert the string representation of lists back into actual lists
spotify_data['artist_name'] = spotify_data['artist_name'].apply(ast.literal_eval)

# Then, extract the first element of each list to get just the artist name
spotify_data['artist_name'] = spotify_data['artist_name'].apply(lambda x: x[0] if x else None)
spotify_data.head()
```

```python
# Merge data
merged_data = pd.merge(spotify_data, aggregated_genres, left_on='artist_name', right_on='artists', how='left')
merged_data.head()
```

```python
def randomly_select_genre(genres):
    if isinstance(genres, list) and genres:
        return random.choice(genres)
    return None

# Apply the function to create a new 'genre' column
merged_data['genre'] = merged_data['genres'].apply(randomly_select_genre)

# Drop the additional columns
merged_data.drop(columns=['artists', 'genres'], inplace=True)
```

```python
# First, ensure that all non-null entries in the 'genre' column are converted from strings to actual lists
merged_data['genre'] = merged_data['genre'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.strip().startswith('[') else x)

# Now, filter out rows where the 'genre' column is an empty list
real_genre_values = merged_data[merged_data['genre'].apply(lambda x: isinstance(x, list) and len(x) > 0)]
print(len(real_genre_values))
```

```python
# Create a mask for rows with non-empty 'genre' lists
non_empty_genre_mask = merged_data['genre'].apply(lambda x: isinstance(x, list) and len(x) > 0)

# For rows with empty 'genre' lists or non-lists, set the 'genre' to 'Unknown'
merged_data.loc[~non_empty_genre_mask, 'genre'] = 'Unknown'
```

```python
# Function to convert any value to a string
def list_to_string(value):
    if isinstance(value, list):
        # Join the list items into a string, separated by commas
        return ', '.join(value)
    return value

# Apply the function to the 'genre' column
merged_data['genre'] = merged_data['genre'].apply(list_to_string)

# Check the DataFrame to see if the changes are applied
print(merged_data['genre'].apply(type).value_counts())
```

### Rearrange columns

```python
# Rearrange columns to make 'song_id' the first column
front_columns = ['song_id', 'song_name', 'artist_id', 'artist_name', 'year', 'release_date']
column_order = front_columns + [col for col in merged_data.columns if col not in front_columns and col != 'popularity'] + ['popularity']
merged_data = merged_data[column_order]

# Confirm changes
merged_data.head()
```
