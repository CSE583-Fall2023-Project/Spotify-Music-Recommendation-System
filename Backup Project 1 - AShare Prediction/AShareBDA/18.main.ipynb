# 数据规整化处理

```python
# 生成 df_bank_toCreateFeature

df_bank = MyDATools.tidyBalanceAndIncomeData(df_BalanceSheet_Bank,df_IncomeStatement_Bank)

#为了分解测试和讲解，存入磁盘
df_bank.to_csv("Data/df_bank_toCreateFeature.csv")


#为了提高运行速度，上一行代码重写
df_bank_toCreateFeature=pd.read_csv("Data/df_bank_toCreateFeature.csv",
                                    index_col= 0, 
                                    converters={'END_DATE': pd.Timestamp})

    #此处，这两个属性是特别重要，否则后续代码出错
        #index_col= 0
        #converters={'END_DATE': pd.Timestamp}
```

# 生成特征矩阵

```python
# 生成 df_bank_toSelectFeature
df_bank_toSelectFeature = MyDATools.createFeatureMatrix(df_bank_toCreateFeature,
                           df_indus,
                           df_comp,
                           df_share,
                           START_DATE=START_DATE,
                           IS_ONEHOT=IS_ONEHOT,
                           N_QUANTILES=N_QUANTILES,
                           N_LAG=N_LAG) 
```

```python
 df_bank_toSelectFeature.to_csv("Data/df_bank_toSelectFeature.csv")
```

# 特征选择

```python
# 生成 df_bank_toTrainModel

df_bank_toTrainModel,df_imp_bank = MyDATools.selectFeatures(df_bank_toSelectFeature,
                                                            P_FEATUREIMP,
                                                            n_estimators=N_TREES,
                                                            max_features=P_FEATRUES)
```

```python
df_bank_toTrainModel.to_csv("Data/df_bank_toTrainModel.csv")
```

# 模型训练

```python
# 生成df_bank_predicted
df_bank_predicted = MyDATools.trainModel(df_bank_toTrainModel,
                        LOG_N=10,
                        etr_EST = 25,
                        etr_FEAT = 0.8,
                        gbr_EST = 200,
                        gbr_FEAT = 0.8,
                        gbr_LRAT = 0.1, 
                        rf_EST = 25,
                        rf_FEAT = 0.8,
                        xgb_EST = 300,
                        xgb_SAM = 0.8,
                        xgb_LRAT =0.1,
                        AlgorithmsInEL=2)
print(df_bank_predicted)
```

```python
#为了分解测试和讲解，暂存预测结果
df_bank_predicted.to_csv("Data/df_bank_predicted.csv")
```

# 模型预测

```python
# 读入竞赛结果df_bank_predicted
df_FDDC_financial_submit = pd.read_csv(PATH_SUBMIT_FILE,
                            sep = '.',
                            header=None,
                            names = ['TICKER_SYMBOL','StockExchange'],
                            dtype = {'TICKER_SYMBOL':str, 'StockExchange':str})

#合并预测结果至FDDC_financial_submit
df_submit_with_predict = MyDATools.appendPredictToSubmit(df_FDDC_financial_submit,
                                                         df_bank_predicted)


#写入预测结果

import datetime
Result_Path = 'Data/submit_'+ datetime.datetime.now().strftime('%Y%m%dT%H')+'.csv'

df_submit_with_predict.to_csv(Result_Path,header=None,index =None)
```
