```python
#此函数的功能
    #[1]对原数据表中的"END_DATE"一列中提取年份和月份
        #END_DATE	End date	截止日期
        #截止日期(END_DATE)：会计期末，比如“2016 年年度报告”中
            #披露的“本期”，END_DATE='2016-12-31'，
            #披露的“上期”，END_DATE='2015-12-31'
            
    #[2]以月份为单位，生成描述月、季度、半年和年份对应关系表
        #DATETIME_UNITS为月份一列代表的时间单位
        
    #[3]将年份和按计算周转换的月份加到原数据表
```

# 读入数据

```python
import pandas as pd
df_BalanceSheet_Bank=pd.read_csv("Data/df_BalanceSheet_Bank.csv",index_col=0)
df_BalanceSheet_Bank.columns = [x.upper() for x in df_BalanceSheet_Bank.columns]
df_BalanceSheet_Bank.describe()
```

```python
df_BalanceSheet_Bank.shape
```

```python
#显式END_DATE一列的值
    #注意：类型为object
df_BalanceSheet_Bank.END_DATE.head(5)
```

# 设置参数

```python
#参数设置
date_column="END_DATE"
DATETIME_UNITS= "QUARTER"
DATETIME_UNITS=DATETIME_UNITS
```

# 将'END_DATE'字段转换为时间类型

```python
#END_DATE列转换为日期类型
df_BalanceSheet_Bank.END_DATE = df_BalanceSheet_Bank['END_DATE'].astype("datetime64[ns]")
df_BalanceSheet_Bank.END_DATE.head()
   #注意：类型变为 datetime64
```

```python
#数据形状没有变
df_BalanceSheet_Bank.shape
```

# 新增year（年份）一列

```python
#提取年份
    #x.year
    
df_BalanceSheet_Bank['year'] = df_BalanceSheet_Bank[date_column].map(lambda x:x.year)
df_BalanceSheet_Bank.year.head()
```

```python
df_BalanceSheet_Bank.shape
```

#  定义月份表——df_month

```python
{'MONTH':df_BalanceSheet_Bank[date_column].astype(str).apply(lambda x:x[5:7])}
```

```python
#提取每个行中的月份
df_month = pd.DataFrame({'MONTH':df_BalanceSheet_Bank[date_column].astype(str).apply(lambda x:x[5:7])})
df_month.head()
```

```python
df_month.shape
```

# 生成月份、季度、年份和半年的对应关系表——df_calendar

```python
df_calendar = pd.DataFrame({'MONTH':['01','02','03','04','05','06','07','08','09','10','11','12'],
                            'QUARTER':['03','03','03','06','06','06','09','09','09','12','12','12'],
                           'HALF':['06','06','06','06','06','06','12','12','12','12','12','12'],
                           'YEAR':['12','12','12','12','12','12','12','12','12','12','12','12']})
df_calendar
```

```python
df_calendar.shape
```

# 合并df_month和 df_calendar

```python
df_month.shape
```

```python
#以月份为依据补充年、季度和半年
df_month_add_calendar = pd.merge(df_month,df_calendar,how='left')
df_month_add_calendar
```

```python
df_month_add_calendar.shape
```
