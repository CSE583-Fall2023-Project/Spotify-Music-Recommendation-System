# 所需调用函数

```python
def changeCycle(df_s,col_date,cycle):
    "依据日期格式生成年月季度半年所在的月"
    df_s['year'] = df_s[col_date].map(lambda x:x.year)
    df_st = pd.DataFrame({'MONTH':df_s[col_date].astype(str).apply(lambda x:x[5:7])})
    df_dict = pd.DataFrame({'MONTH':['01','02','03','04','05','06','07','08','09','10','11','12'],
                            'QUARTER':['03','03','03','06','06','06','09','09','09','12','12','12'],
                           'HALF':['06','06','06','06','06','06','12','12','12','12','12','12'],
                           'YEAR':['12','12','12','12','12','12','12','12','12','12','12','12']})
    df_output = pd.merge(df_st,df_dict,how='left')
    df_s = df_s.reset_index()
    df_s['month'] = df_output[cycle]
    return(df_s)
```

# 变量设置

```python
#先看一下变化之前的数据
PATH_MACROINDUSTRY_FILE = 'Data/MacroIndustry.xlsx'
filename=PATH_MACROINDUSTRY_FILE
```

# 读取MacroIndustry的sheet ——INDIC_DATA 

```python
import pandas as pd
df_indus = pd.DataFrame(pd.read_excel(filename,sheet_name=['INDIC_DATA'])['INDIC_DATA'])
df_indus.head()
```

```python
#字段名称统一改为大写
df_indus.columns = [x.upper() for x in df_indus.columns]
df_indus.head()
```

# 新增 year和month两个列

```python
#原表中新增 year 和 mongth两个列
    #调用函数changeCycle（）
CYCLE = "QUARTER"#YEAR，QUARTER，HALF
df_indus1 = changeCycle(df_indus,'PERIOD_DATE',CYCLE)
df_indus1.head()
```

# 定义分组统计表

```python
 #生成分组统计df_indus_g
    #按'INDIC_ID','year','month（季度）'进行分组统计
    #统计内容为DATA_VALUE
df_indus_g =  df_indus1.groupby(['INDIC_ID','year','month'])[['DATA_VALUE']].mean().reset_index()
df_indus_g['INDIC_ID'] = df_indus_g['INDIC_ID'].astype(str)
df_indus_g.head()
```

```python
df_indus_g.shape
```

# 生成透视表（主键为year+month）

```python
#生成透视表
    #主键为 year+month
    #列名为上表的INDIC_ID
        #宏观统计指标
    #值为DATA_VALUE
df_indus_g1 = df_indus_g.pivot_table(index=['year','month'],
                                     columns=["INDIC_ID"],
                                     values=["DATA_VALUE"]).reset_index()
df_indus_g1.head()
```

```python
#查看表的列名
df_indus_g1.columns
    #注意：df_indus_g1结果为多级索引
```

```python
#将上表（df_indus_g1）的表头合并成一行
    #df_indus_g1带有二级索引
    
df_indus_g1.columns = [ ''.join(col) for col in df_indus_g1.columns.values]
df_indus_g1.head()
```

```python
#填充缺失值
df_indus_g1 = df_indus_g1.fillna(method='bfill')
df_indus_g1.head()
```
