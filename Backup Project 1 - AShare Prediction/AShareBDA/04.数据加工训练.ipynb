# 4.数据加工训练

## 4.1读入数据并查看数据

```python
#导入数据包
import numpy as np
import pandas as pd
```

```python
#读入并查看数据

#读入数据
dtype={'TICKER_SYMBOL':str}  #按规定格式读取某列
df1 = pd.read_csv(\
                  "Data/GeneralBusinessIncome.csv",\
                  dtype = dtype)
df2 = pd.read_csv("Data/BankIncome.csv",\
                  dtype = dtype)

#查看数据
df1
df2
```

## 4.2 数据变换——大小变换之min-max（0-1）标准化

```python
from sklearn.preprocessing import MinMaxScaler
#建立MinMaxScaler对象
mmscaler=MinMaxScaler()
#标准化处理
data_mms =mmscaler.fit_transform(df1['REVENUE'].values.reshape(-1,1))
data_mms
```

## 4.3 数据变换——大小变换之Z-score 标准化

```python
from sklearn.preprocessing import StandardScaler

#建立StandardScaler对象
sscaler = StandardScaler()

#标准化处理
data_zs = sscaler.fit_transform(df1['REVENUE'].values.reshape(-1,1))
data_zs
```

## 4.4 数据变换——大小变换之normalizer 归一化

```python
from sklearn.preprocessing import Normalizer

#建立Normalizer对象
nscaler=Normalizer()

#删去缺失值
data_del = df1.drop(labels = df1.index[df1['T_REVENUE'].isnull()],axis=0)  

#标准化处理
data_ns = nscaler.fit_transform(data_del['T_REVENUE'].values.reshape(1,-1))   
data_ns
```

```python
# 数据变换——大小变换(汇总)
from sklearn import preprocessing

# # min-max（0-1）标准化
mmscaler=preprocessing.MinMaxScaler()
data_mms = mmscaler.fit_transform(df1['REVENUE'].values.reshape(-1,1))

# Z-score 标准化
sscaler = preprocessing.StandardScaler()
data_zs = sscaler.fit_transform(df1['REVENUE'].values.reshape(-1,1))

# normalizer 归一化
nscaler=preprocessing.Normalizer()
data_del = df1.drop(labels = df1.index[df1['T_REVENUE'].isnull()],axis=0)  #删去缺失值
data_ns = nscaler.fit_transform(data_del['T_REVENUE'].values.reshape(1,-1))   

data_list = [data_mms,data_zs,data_ns]
data_list
```

## 4.8 数据排序——索引排序

```python
# 对行索引，进行降序排列
data_sort11 = df1.sort_index(axis=0,ascending=False,inplace=False)
data_sort11

# 对列索引，进行升序排列
data_sort12 = df1.sort_index(axis=1,ascending=True,inplace=False)
data_sort12
```

## 4.9 数据排序——值排序

```python
#对某一列进行降序排列
data_sort21 = df1.sort_values(\
                              by="REVENUE",axis=0,ascending=False,inplace=False)
data_sort21['REVENUE']

#对多列进行联合排序
data_sort22 = df1.sort_values(\
                              by=["REVENUE","T_REVENUE"],\
                              axis=0,ascending=[True,True],\
                              inplace=False,na_position="last")
data_sort22
```

## 4.10 数据排序——值排名

```python
# method="first"
data_torank = df1.iloc[:10,[1,10]]
data_torank
data_torank['RANK'] = data_torank["REVENUE"].rank(method="first")
data_torank  #指定的排名列如果出现相同的值，不同的method计算排名方法不同
```
