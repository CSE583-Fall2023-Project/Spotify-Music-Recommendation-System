```python
# unique_songs = data.drop_duplicates(subset=['song_id']).reset_index(drop=True)[['song_id', 'song_name']]
unique_songs = data.drop_duplicates(subset=['song_name']).reset_index(drop=True)[['song_id', 'song_name']]
unique_songs
```

```python
df_songs_features = data.pivot_table(index='song_id', columns='user_id', values='listening_count').fillna(0)
df_songs_features = df_songs_features.loc[unique_songs.values[:, 0]]
```

```python
# obtain a sparse matrix
mat_songs_features = csr_matrix(df_songs_features.values)
```

```python
decode_id_song = {
    song: i for i, song in 
    enumerate(list(unique_songs.set_index('song_id').loc[df_songs_features.index].song_name))
}
```

```python
from knn_recommender import Recommender
```

```python
model = Recommender(metric='cosine', algorithm='brute', k=20, data=mat_songs_features, decode_id_song=decode_id_song)
```

```python
song = 'I believe in miracles'
# song = 'Banana Republics'
decode_id_song
```

```python
new_recommendations = model.make_recommendation(new_song=song, n_recommendations=10)
```

```python
print(f"The recommendations for {song} are:")
print(f"{new_recommendations}")
```
