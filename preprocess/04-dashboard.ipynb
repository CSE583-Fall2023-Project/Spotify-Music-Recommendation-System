```python
# Import the liraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load the data
spotify_data = pd.read_csv("../data/spotify_data.csv")

# Preview the data
spotify_data.head()
```

```python
# Load the data
user_songs = pd.read_csv("../data/user_songs.csv")

# Preview the data
user_songs.head()
```

```python
# Filter the user_songs dataframe for the specific user_id
user_id = "19678c5c-9d86-4c9e-a0e4-4fa173a8bb5a"
filtered_user_songs = user_songs[user_songs['user_id'] == user_id]

# Inner join the data frames on song_id
joined_df = pd.merge(filtered_user_songs, spotify_data, on = "song_id", how = "inner")

# Select the required columns
selected_columns = [
    "song_id", "listening_count", "song_name", "artist_name", "year", 
    "valence", "acousticness", "danceability", "energy", "instrumentalness", 
    "liveness", "speechiness"
]
result_df = joined_df[selected_columns]

result_df.head()  # Display the first few rows of the result dataframe
```

```python
import numpy as np

# Extracting the columns for the weighted average calculation
attributes = ["valence", "acousticness", "danceability", "energy", "instrumentalness", "liveness", "speechiness"]
weights = result_df["listening_count"].values

# Calculating the weighted average for each attribute
weighted_averages = {}
for attribute in attributes:
    values = result_df[attribute].values
    weighted_average = np.sum(values * weights) / np.sum(weights)
    weighted_averages[attribute] = weighted_average

# Creating a DataFrame for the result
weighted_averages_df = pd.DataFrame(weighted_averages, index=[0])
weighted_averages_df
```
